### Helper

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'sans-serif' 
plt.rcParams['font.sans-serif'] = 'DejaVu Serif' 
```

```python
def parse_one_result(one_result):
    one_result_copy = one_result.copy()
    normal_acc, contrast_acc = [], []
    for i in range(5):
        a, b = one_result_copy["contrast_rh{}".format(i)]
        one_result_copy.pop("contrast_rh{}".format(i))
        normal_acc.append(a)
        contrast_acc.append(b)

    one_result_copy["contrast_rh"] = (np.mean(normal_acc), np.mean(contrast_acc))
    one_result_copy["contrast_rh_std"] = (np.std(normal_acc), np.std(contrast_acc))

    y1 = [one_result_copy[key][0] for key in ["base", "contrast", "contrast_fih", "contrast_rh"]]
    y2 = [one_result_copy[key][1] for key in ["base", "contrast", "contrast_fih", "contrast_rh"]]

    std1 = [0, 0, 0] + [one_result_copy["contrast_rh_std"][0]]
    std2 = [0, 0, 0] + [one_result_copy["contrast_rh_std"][1]]

    return y1, std1, y2, std2
    
```

```python
labels = ["Base Prompt", "Contrast Prompt", "Contrast Prompt\n(Ablate 6 FI Heads)", "Contrast Prompt\n(Ablate 6 Rand. Heads)"]
def plot_one_result(ax, one_result, title, yticks=False, legend=False, baseline=False, labels=labels):
    y1, std1, y2, std2 = parse_one_result(one_result)
    x = np.arange(len(y1))  # the label locations
    width = 0.4  # the width of the bars

    rects2 = ax.barh(x + width/2, y1[::-1], width, xerr=std1[::-1], label='Base Acc', color="#bf3100")
    rects1 = ax.barh(x - width/2, y2[::-1], width, xerr=std2[::-1], label='Contrast Acc', color="#ec9f05")

    # Add labels, title, and legend
    ax.set_xlabel("Accuracy")
    ax.set_yticks(x)
    if baseline:
        ax.axvline(x=y2[0], linewidth=1.0, color="lightgray", linestyle=":")
    if yticks:
        ax.set_yticklabels(labels[::-1])
    else:
        ax.set_yticklabels([""] * 4)
    ax.set_xticks([0, 0.5, 1.0])
    ax.set_xticklabels(["0%", "50%", "100%"])
    if legend:
        ax.legend(bbox_to_anchor=(1.8, 0.5), loc='center right', ncol=1, fontsize=9)
    ax.set_xlim(0, 1)
    ax.set_title(title, fontsize=10)
```
